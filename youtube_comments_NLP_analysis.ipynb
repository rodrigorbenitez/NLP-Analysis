# Conversión del texto en string

```python
# Function to convert   
def listToString(s):  
    
    # initialize an empty string 
    str1 = ""  
    
    # traverse in the string   
    for ele in s:  
        str1 += ele   
    
    # return string   
    return str1
```

```python
output_text=listToString(output)
```

# Bloque de texto total en string

```python
output_text
```

# Eliminación de saltos de linea y reemplazo con espacio

```python
output_text = output_text.replace('\n', ' ')
```

# Reemplazo de mas de un espacio en el texto con un solo espacio

```python
import re
output_text =re.sub(' +', ' ', output_text)
output_text=output_text.lower()
```

# Texto final sin saltos de linea y sin doble espacios

```python
output_text
```

# Eliminación de links a paginas web

```python
output_text=re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", output_text)
```

# Carga del modulo en español de Spacy

```python
nlp= spacy.load("es_core_news_sm")
```

# NLP (Natural language processing) con spaCy

```python
doc = nlp(output_text)
```

# Extraccion de sintagmas nominales

```python
key_concepts=[]
```

```python
key_concepts=[chunk.text for chunk in doc.noun_chunks]
    
```

# Creacion de listado para extraccion de nombres de entidades

```python
# Find named entities, phrases and concepts
entidades=[]
for entity in doc.ents:
    entidades.append(entity.text+","+entity.label_)
```

```python
entidades
```

```python
df_entidades = DataFrame (entidades,columns=['entidad'])
```

```python
df_entidades
```

```python
df_entidades.to_csv('entidades.csv')
```

# Extraccion de verbos

```python
verbos=[token.lemma_ for token in doc if token.pos_ == "VERB"]
```
